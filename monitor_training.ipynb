```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def moving_average(array, N=50):
    padded_array = np.pad(array, (N//2, N-1-N//2), mode='edge')
    return np.convolve(padded_array, np.ones((N,))/N, mode='valid')
```

```python
def plot_loss(train_x, train_y, test_x, test_y, label, training_fraction=0.9):
    plt.plot(train_x, train_y, label = 'training')
    plt.plot(test_x, test_y, label = 'test')
    plt.legend()
    plt.xlabel("Training step")
    plt.ylabel("Loss")
    plt.title(label)
    plt.show()
    
```

```python
log_dir = './logs'
label = 'JUNIPR_test' # label to loss file in log folder
```

```python
test_loss  = np.loadtxt(log_dir + '/' + label + '_test_loss.log')
train_loss = np.loadtxt(log_dir + '/' + label + '_train_loss.log')
```

```python
train_x = train_loss[:,0]
test_x = test_loss[:,0]

train1 = moving_average(train_loss[:,1])
test1  = test_loss[:,1]

train2 = moving_average(train_loss[:,2])
test2  = test_loss[:,2]

train3 = moving_average(train_loss[:,3])
test3  = test_loss[:,3]

train4 = moving_average(train_loss[:,4])
test4  = test_loss[:,4]
```

```python
plot_loss(train_x, train1, test_x, test1, "Total Loss")
```

```python
plot_loss(train_x, train2, test_x, test2, "End Loss")
```

```python
plot_loss(train_x, train3, test_x, test3, "Mother Loss")
```

```python
plot_loss(train_x, train4, test_x, test4, "Branch Loss")
```
